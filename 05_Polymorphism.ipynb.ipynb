# Polymorphism


Python me **polymorphism** ka matlab hai **ek hi interface ya method name ka alag-alag behavior**.  
- Iska use program me **flexibility aur reusability** badhane ke liye hota hai.  
- Polymorphism do tarah se implement hota hai:  
  1. **Method Overriding** → Child class parent class ka method override karta hai.  
  2. **Operator/Function Overloading (Python me dynamic typing se achieve hota hai)** → Same operator ya function different types ke saath behave kar sakta hai.  

---

- **Method Overriding**: Child class me parent class ka method redefine karna.

- **Operator Overloading**: Same operator different objects pe different behavior show kare.




```python
# Example 1: Method Overriding
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child(Parent):
    def greet(self):
        print("Hello from Child")

p = Parent()
c = Child()
p.greet()  # Parent method
c.greet()  # Child method overrides Parent


```

```python

# Example 2: Polymorphism with different classes
class Dog:
    def sound(self):
        print("Woof!")

class Cat:
    def sound(self):
        print("Meow!")

animals = [Dog(), Cat()]
for animal in animals:
    animal.sound()  # Same method name, different behavior

```

```python
# Example 3: Function with dynamic typing
def add(a, b):
    return a + b

print(add(5, 10))      # Integer addition
print(add("Hi, ", "Bob"))  # String concatenation
print(add([1,2], [3,4]))   # List concatenation


# Example 4: Operator polymorphism
print(10 + 20)      # Integer addition
print("Hello " + "World")  # String concatenation
print([1,2] + [3,4])       # List concatenation


# Example 5: Polymorphism with inheritance
class Shape:
    def area(self):
        print("Area not defined")

class Square(Shape):
    def __init__(self, side):
        self.side = side
    def area(self):
        print("Square area:", self.side ** 2)

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        print("Circle area:", 3.14 * self.radius ** 2)

shapes = [Square(4), Circle(3)]
for shape in shapes:
    shape.area()  # Same method name, different behavior


```
