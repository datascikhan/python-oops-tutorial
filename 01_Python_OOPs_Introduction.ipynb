
# Python OOPs (Object-Oriented Programming) 


Python me **OOPs (Object-Oriented Programming)** ka use real-world entities ko **objects ke form me represent** karne ke liye hota hai.  
- Is approach me **data (attributes)** aur **functions (methods)** ko ek unit me **class** ke andar organize kiya jata hai.  
- Objects, classes ke **instances** hote hain aur ye real-world entities ka behavior aur properties carry karte hain.  
- OOPs ka main goal hai **code reusability, modularity, aur abstraction** provide karna, taake complex programs ko manage karna easy ho.  











**Key Concepts:**  
- **Class:** Blueprint ya template jo object ka structure define karta hai.  
- **Object:** Class ka instance, real-world entity ka representation.  
- **Attributes:** Object ke data/properties.  
- **Methods:** Object ke behavior/functions.  
- **Encapsulation:** Data aur methods ko ek unit me combine karna.  
- **Inheritance:** Existing class ke features ko dusri class me reuse karna.  
- **Polymorphism:** Same method name ka alag-alag behavior.  
- **Abstraction:** Complex details ko hide karke sirf important features show karna.


# 1️⃣ Basic Concepts of OOPs

| Concept       | Description                                              |
| ------------- | -------------------------------------------------------- |
| Class         | Blueprint ya template for objects                        |
| Object        | Instance of a class                                      |
| Attributes    | Variables inside class (state)                           |
| Methods       | Functions inside class (behavior)                        |
| Inheritance   | Ek class dusri class se properties inherit kar sakti hai |
| Encapsulation | Data ko secure rakhna (private/protected)                |
| Polymorphism  | Same method different ways me behave karna               |
| Abstraction   | Internal implementation hide karna                       |
