#  Creating a Class and Object


Python me **class** ek blueprint hai jo **objects** banane ke liye use hoti hai.  
**Object** class ka instance hota hai aur ye class ke **attributes** aur **methods** ka use kar sakta hai.  
Class me **constructor (`__init__`)** define karke object ke initial attributes set kiye ja sakte hain.



```python
# Example 1: Simple class and object
class Person:
    pass  # Empty class

p1 = Person()  # Creating an object
print("Object created:", p1)
```

```python

# Example 2: Class with attributes
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Alice", 25)
print("Name:", p1.name)
print("Age:", p1.age)
```

```python
# Example 3: Class with method
class Person:
    def __init__(self, name):
        self.name = name
    
    def greet(self):
        print("Hello, my name is", self.name)

p1 = Person("Bob")
p1.greet()  # Calling method
```

```python
# Example 4: Multiple objects from same class
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

c1 = Car("Toyota", "Corolla")
c2 = Car("Honda", "Civic")

print("Car 1:", c1.brand, c1.model)
print("Car 2:", c2.brand, c2.model)
```

```python
# Example 5: Updating object attributes
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Charlie", 30)
print("Before update:", p1.name, p1.age)

# Updating attributes
p1.name = "David"
p1.age = 35
print("After update:", p1.name, p1.age)
```


# Constructor (`__init__`)

Python me **constructor** ek special method hai jo **automatically call hota hai** jab bhi class ka object create hota hai.  
- Iska naam hamesha `__init__` hota hai.  
- Constructor ka use **object ke initial attributes** set karne ke liye hota hai.  
- Ye object ke saath **automatic initialization** provide karta hai.




```python

# Example 1: Basic constructor
class Person:
    def __init__(self, name):
        self.name = name

p1 = Person("Alice")
print("Name:", p1.name)
```

```python
# Example 2: Constructor with multiple attributes
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = Person("Bob", 25)
print("Name:", p1.name, "Age:", p1.age)
```

```python

# Example 3: Constructor with default values
class Person:
    def __init__(self, name="Unknown", age=0):
        self.name = name
        self.age = age

p1 = Person()
print("Name:", p1.name, "Age:", p1.age)
```

```python
# Example 4: Using constructor to initialize a list
class Classroom:
    def __init__(self, students):
        self.students = students

class1 = Classroom(["Alice", "Bob", "Charlie"])
print("Students:", class1.students)
```

```python

# Example 5: Constructor calling a method
class Person:
    def __init__(self, name):
        self.name = name
        self.greet()  # Call a method inside constructor

    def greet(self):
        print("Hello, my name is", self.name)

p1 = Person("David")
```
