# Inheritance

Python me **inheritance** ek mechanism hai jisme ek class (child/subclass) **dusri class (parent/superclass) ke attributes aur methods** inherit kar sakti hai.  
- Iska use **code reuse** aur **hierarchical relationships** create karne ke liye hota hai.  
- Child class parent class ke features ko extend ya modify kar sakti hai.  

---


```python
# Example 1: Single inheritance
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child(Parent):
    pass

c1 = Child()
c1.greet()  # Inherited method
```

```python
# Example 2: Child class adding its own method
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child(Parent):
    def play(self):
        print("Child is playing")

c1 = Child()
c1.greet()
c1.play()
```

```python

# Example 3: Constructor inheritance
class Parent:
    def __init__(self, name):
        self.name = name

class Child(Parent):
    def display(self):
        print("Name:", self.name)

c1 = Child("Alice")
c1.display()
```

```python


# Example 4: Overriding parent method
class Parent:
    def greet(self):
        print("Hello from Parent")

class Child(Parent):
    def greet(self):
        print("Hello from Child")

c1 = Child()
c1.greet()  # Child method overrides parent
```

```python
# Example 5: Using super() to call parent constructor
class Parent:
    def __init__(self, name):
        self.name = name

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)  # Call parent constructor
        self.age = age
    
    def display(self):
        print("Name:", self.name, "Age:", self.age)

c1 = Child("Bob", 20)
c1.display()

```
